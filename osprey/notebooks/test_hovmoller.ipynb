```python
import matplotlib.pyplot as plt
```

```python
from osprey.graphics.hovmoller import hovmoller
```

```python
hovmoller(expname='lfr0', startyear=1990, endyear=2010, varlabel='thetao', format="plain", rescale=True)
```

```python
hovmoller(expname='FE01', startyear=1990, endyear=2099, varname='thetao', reader='post', metric='diff', rescale=True)
plt.title('metric=diff')
```

```python
hovmoller(expname='FE01', startyear=1990, endyear=2099, varname='thetao', reader='post', metric='reldiff', rescale=True)
plt.title('metric=relative difference')
```
