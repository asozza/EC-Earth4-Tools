```python
import matplotlib.pyplot as plt
```

```python
from osprey.graphics.hovmoller import hovmoller
```

```python
hovmoller(expname='lfr0', startyear=1990, endyear=2030, varlabel='thetao', format="plain", replace=True, rescale=True)
```

```python
refinfo = {'expname': 'lgr3', 'startyear': 2340, 'endyear': 2349, 'diagname': 'profile', 'format': 'global'}
hovmoller(expname='FE01', startyear=2220, endyear=2230, varlabel='thetao', replace=True, metric='diff', refinfo=refinfo, rescale=False)
```

```python
hovmoller(expname='FE01', startyear=1990, endyear=2099, varname='thetao', reader='post', metric='reldiff', rescale=True)
plt.title('metric=relative difference')
```
