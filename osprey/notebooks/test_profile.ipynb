```python
import xarray as xr
import matplotlib.pyplot as plt
```

```python
from osprey.graphics.profile import profile
```

```python
from osprey.actions.reader import reader_nemo_field
```

```python
data = reader_nemo_field(expname='lfr0', startyear=2220, endyear=2230, varname='woto')
```

```python
from osprey.actions.postreader import averaging
```

```python
ave = averaging(data=data, varlabel='woto', diagname='profile', format='global', orca='ORCA2')
```

```python
plt.plot(ave['z'],ave['woto'])
```

```python
profile(expname='lfr0', startyear=2220, endyear=2230, varlabel='woto', reader='post', replace=True)
profile(expname='lgr4', startyear=2340, endyear=2349, varlabel='woto', reader='post', replace=True)
```
