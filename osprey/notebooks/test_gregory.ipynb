```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
```

```python
from osprey.graphics.gregory_plot import gregory_plot
```

```python
gregory_plot(expname='lfr0', startyear=1990, endyear=2079, varname1='thetao', varname2='qt_oce', reader='post', metric='base', color='lightcoral', linestyle='--', label='REF')
gregory_plot(expname='FE01', startyear=1990, endyear=2079, varname1='thetao', varname2='qt_oce', reader='post', metric='base', color='red', linestyle='-', label='EOF')
plt.legend(
    bbox_to_anchor=(0.98, 0.98),  # x, y coordinates for legend placement
    loc='upper right',         # Location of the legend relative to bbox_to_anchor
    borderaxespad=0           # Padding between the legend and the plot
)
plt.title('Gregory plot')
```

```python
gregory_plot(expname='lfr0', startyear=1990, endyear=2079, varname1='tos', varname2='qt_oce', reader='post', metric='base', color='darkseagreen', linestyle='--', label='REF')
gregory_plot(expname='FE01', startyear=1990, endyear=2079, varname1='tos', varname2='qt_oce', reader='post', metric='base', color='green', linestyle='-', label='EOF')
plt.legend(
    bbox_to_anchor=(0.98, 0.98),  # x, y coordinates for legend placement
    loc='upper right',         # Location of the legend relative to bbox_to_anchor
    borderaxespad=0           # Padding between the legend and the plot
)
plt.title('Gregory plot')
```
