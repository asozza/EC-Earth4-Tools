```python
# Create a figure with 3 subplots (one for each layer)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Surface layer
timeseries(expname='lfr0', startyear=1990, endyear=2079, varlabel='tos', reader='post', 
           rescale=False, color='lightcoral', linestyle='--', label='surface - REF', ax=axes[0])
timeseries(expname='FE01', startyear=1990, endyear=2079, varlabel='tos', reader='post', 
           rescale=False, color='red', linestyle='-', label='surface - EOF', ax=axes[0])
axes[0].legend(loc='upper right')
axes[0].set_title('Surface Layer')

# Pycnocline layer
timeseries(expname='lfr0', startyear=1990, endyear=2079, varlabel='thetao-pyc', reader='post', 
           rescale=False, color='darkseagreen', linestyle='--', label='pycnocline - REF', ax=axes[1])
timeseries(expname='FE01', startyear=1990, endyear=2079, varlabel='thetao-pyc', reader='post', 
           rescale=False, color='green', linestyle='-', label='pycnocline - EOF', ax=axes[1])
axes[1].legend(loc='upper right')
axes[1].set_title('Pycnocline Layer')

# Bottom layer
timeseries(expname='lfr0', startyear=1990, endyear=2079, varlabel='sbt', reader='post', 
           rescale=False, color='cornflowerblue', linestyle='--', label='bottom - REF', ax=axes[2])
timeseries(expname='FE01', startyear=1990, endyear=2079, varlabel='sbt', reader='post', 
           rescale=False, color='blue', linestyle='-', label='bottom - EOF', ax=axes[2])
axes[2].legend(loc='upper right')
axes[2].set_ylabel('Temperature [degC]')
axes[2].set_xlabel('Time [years]')

# Adjust layout to prevent overlap
plt.tight_layout()

# Set the main title for the figure

fig.suptitle('Temperature by Ocean Layer', fontsize=16, y=1.02)
```

```python
timeseries(expname='lfr0', startyear=1990, endyear=2020, varlabel='thetao', reader='post', metric='diff', rescale=True, color='lightcoral', linestyle='--', label='diff - REF')
timeseries(expname='FE01', startyear=1990, endyear=2020, varlabel='thetao', reader='post', metric='diff', rescale=True, color='red', linestyle='-', label='diff - EOF T')
timeseries(expname='FE02', startyear=1990, endyear=2020, varlabel='thetao', reader='post', metric='diff', rescale=True, color='orange', linestyle='-', label='diff - EOF TS')
plt.legend(
    bbox_to_anchor=(0.98, 0.98),  # x, y coordinates for legend placement
    loc='upper right',         # Location of the legend relative to bbox_to_anchor
    borderaxespad=0           # Padding between the legend and the plot
)
plt.ylabel('C(t)/C(0) []')
plt.xlabel('t [years]')
plt.title("Cost functions of global mean temperature \n with respect to the equilibrium state [2390-2400]")
```

```python
timeseries(expname='lfr0', startyear=1990, endyear=2020, varlabel='so', reader='post', metric='diff', rescale=True, color='lightcoral', linestyle='--', label='diff - REF')
timeseries(expname='FE01', startyear=1990, endyear=2020, varlabel='so', reader='post', metric='diff', rescale=True, color='red', linestyle='-', label='diff - EOF T')
timeseries(expname='FE02', startyear=1990, endyear=2020, varlabel='so', reader='post', metric='diff', rescale=True, color='orange', linestyle='-', label='diff - EOF TS')
plt.legend(
    bbox_to_anchor=(0.98, 0.98),  # x, y coordinates for legend placement
    loc='upper right',         # Location of the legend relative to bbox_to_anchor
    borderaxespad=0           # Padding between the legend and the plot
)
plt.ylabel('C(t)/C(0) []')
plt.xlabel('t [years]')
plt.title("Cost functions of global mean salinity \n with respect to the equilibrium state [2390-2400]")
```

```python
# FE01, FE02

# Create a figure with 3 subplots (one for each layer)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Surface layer
timeseries(expname='lfr0', startyear=1990, endyear=2020, varlabel='tos', reader='post', 
           rescale=False, color='lightcoral', linestyle='--', label='surface - REF', ax=axes[0])
timeseries(expname='FE01', startyear=1990, endyear=2020, varlabel='tos', reader='post', 
           rescale=False, color='red', linestyle='-', label='surface - EOF-T', ax=axes[0])
timeseries(expname='FE02', startyear=1990, endyear=2020, varlabel='tos', reader='post', 
           rescale=False, color='orange', linestyle='-', label='surface - EOF-TS', ax=axes[0])
axes[0].legend(loc='upper right')
axes[0].set_title('Surface Layer')

# Pycnocline layer
timeseries(expname='lfr0', startyear=1990, endyear=2020, varlabel='thetao-pyc', reader='post', 
           rescale=False, color='darkseagreen', linestyle='--', label='pycnocline - REF', ax=axes[1])
timeseries(expname='FE01', startyear=1990, endyear=2020, varlabel='thetao-pyc', reader='post', 
           rescale=False, color='green', linestyle='-', label='pycnocline - EOF-T', ax=axes[1])
timeseries(expname='FE02', startyear=1990, endyear=2020, varlabel='thetao-pyc', reader='post', 
           rescale=False, color='lime', linestyle='-', label='pycnocline - EOF-TS', ax=axes[1])
axes[1].legend(loc='upper right')
axes[1].set_title('Pycnocline Layer')

# Bottom layer
timeseries(expname='lfr0', startyear=1990, endyear=2020, varlabel='sbt', reader='post', 
           rescale=False, color='cornflowerblue', linestyle='--', label='bottom - REF', ax=axes[2])
timeseries(expname='FE01', startyear=1990, endyear=2020, varlabel='sbt', reader='post', 
           rescale=False, color='blue', linestyle='-', label='bottom - EOF-T', ax=axes[2])
timeseries(expname='FE01', startyear=1990, endyear=2020, varlabel='sbt', reader='post', 
           rescale=False, color='deepskyblue', linestyle='-', label='bottom - EOF-TS', ax=axes[2])
axes[2].legend(loc='upper right')
axes[2].set_ylabel('Temperature [degC]')
axes[2].set_xlabel('Time [years]')

# Adjust layout to prevent overlap
plt.tight_layout()

# Set the main title for the figure
fig.suptitle('Temperature by Ocean Layer', fontsize=16, y=1.02)
```
