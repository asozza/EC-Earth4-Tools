```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
```

```python
from osprey.graphics.timeseries import timeseries, timeseries_two
```

```python
timeseries(expname='lfr0', startyear=1990, endyear=2079, varlabel='thetao', reader='post', metric='diff', rescale=True, color='lightcoral', linestyle='--', label='diff - REF')
timeseries(expname='lfr0', startyear=1990, endyear=2079, varlabel='thetao', reader='post', metric='reldiff', rescale=True, color='bisque', linestyle='--', label='reldiff - REF')
timeseries(expname='lfr0', startyear=1990, endyear=2079, varlabel='thetao', reader='post', metric='abs', rescale=True, color='khaki', linestyle='--', label='abs - REF')
timeseries(expname='lfr0', startyear=1990, endyear=2079, varlabel='thetao', reader='post', metric='relabs', rescale=True, color='darkseagreen', linestyle='--', label='relabs - REF')
timeseries(expname='lfr0', startyear=1990, endyear=2079, varlabel='thetao', reader='post', metric='sqerr', rescale=True, color='cornflowerblue', linestyle='--', label='sqerr - REF')
timeseries(expname='lfr0', startyear=1990, endyear=2079, varlabel='thetao', reader='post', metric='relsqerr', rescale=True, color='thistle', linestyle='--', label='relsqerr - REF')
#
timeseries(expname='FE01', startyear=1990, endyear=2079, varlabel='thetao', reader='post', metric='diff', rescale=True, color='red', linestyle='-', label='diff - EOF')
timeseries(expname='FE01', startyear=1990, endyear=2079, varlabel='thetao', reader='post', metric='reldiff', rescale=True, color='orange', linestyle='-', label='reldiff - EOF')
timeseries(expname='FE01', startyear=1990, endyear=2079, varlabel='thetao', reader='post', metric='abs', rescale=True, color='brown', linestyle='-', label='abs - EOF')
timeseries(expname='FE01', startyear=1990, endyear=2079, varlabel='thetao', reader='post', metric='relabs', rescale=True, color='green', linestyle='-', label='relabs - EOF')
timeseries(expname='FE01', startyear=1990, endyear=2079, varlabel='thetao', reader='post', metric='sqerr', rescale=True, color='blue', linestyle='-', label='sqerr - EOF')
timeseries(expname='FE01', startyear=1990, endyear=2079, varlabel='thetao', reader='post', metric='relsqerr', rescale=True, color='purple', linestyle='-', label='relsqerr - EOF')
plt.legend(
    bbox_to_anchor=(1.05, 1),  # x, y coordinates for legend placement
    loc='upper left',         # Location of the legend relative to bbox_to_anchor
    borderaxespad=0           # Padding between the legend and the plot
)
plt.ylabel('C(t)/C(0) []')
plt.xlabel('t [years]')
plt.title("Cost functions of global mean temperature \n with respect to the equilibrium state [2390-2400]")
```
