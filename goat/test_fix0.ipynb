```python
vec = {}
expname='fix0'
dirs = io.folders(expname)
data = io.readmf_T(expname, 1990, 2000)
df = gm.elements(expname)
vec[0] = gt.dateDecimal(data['time'].values)
```

```python
vec[1] = gm.movave(data['thetao'].weighted(df['vol']).mean(dim=['z', 'y', 'x']).values.flatten(),12)
```

```python
vec[2] = gm.movave(data['qt_oce'].weighted(df['vol']).mean(dim=['z', 'y', 'x']).values.flatten(),12)
```

```python
with netCDF4.Dataset(os.path.join("/perm/itas/ece4", expname, "nemo", 'output.nc'), 'w', format='NETCDF4') as ncout:
    # create time dimension
    nhours = len(vec[0])
    time = ncout.createDimension('time', nhours)

    # create the time variable
    years = ncout.createVariable('time', 'f8', ('time',))
    years.units = 'decimalYear'
    years.calendar = "gregorian"
    years[:] = vec[0][:]

    # create variable 'thetao', dependent on time
    theta = ncout.createVariable('thetao', 'f8', ('time',))
    theta.units = "degC"
    theta.coordinates = "time"
    theta[:] = vec[1][:]

    # create variable 'qt_oce'
    qtoce = ncout.createVariable('qt_oce', 'f8', ('time',))
    qtoce.units = "W/m2" 
    qtoce.coordinates = "time"
    qtoce[:] = vec[2][:]
```

```python
vec[3] = gm.movave(data['tos'].weighted(df['area']).mean(dim=['y', 'x']).values.flatten(),12)
```

```python
# append another variables
with netCDF4.Dataset(os.path.join("/perm/itas/ece4", expname, "nemo", 'output.nc'), 'a', format='NETCDF4') as ncout:
    # create variable 'qt_oce'
    sst = ncout.createVariable('tos', 'f8', ('time',))
    sst.units = "degC" 
    sst.coordinates = "time"
    sst[:] = vec[3][:]
```
